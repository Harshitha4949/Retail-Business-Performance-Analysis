```python
category_sales = df.groupby("Category")["Sales"].sum()

category_sales
```

```python
category_profit = df.groupby("Category")["Profit"].sum()

category_profit
```

```python
region_sales = df.groupby("Region")["Sales"].sum()

region_sales
```

```python
region_profit = df.groupby("Region")["Profit"].sum()

region_profit
```

```python
df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10)
```

```python
category_sales.plot(kind="bar")

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.show()
```

```python
category_profit.plot(kind="bar")

plt.title("Profit by Category")

plt.show()
```

```python
df.groupby("Segment")["Sales"].sum().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.ylabel("")
plt.title("Sales by Segment")

plt.show()
```

```python
df.groupby("State")["Sales"].sum().sort_values().tail(10).plot(kind="barh")

plt.title("Top 10 States by Sales")

plt.show()
```

```python
plt.scatter(df["Sales"], df["Profit"])

plt.xlabel("Sales")
plt.ylabel("Profit")
plt.title("Sales vs Profit")

plt.show()
```

```python
print("Highest Sales Category :", category_sales.idxmax())
print("Highest Profit Category :", category_profit.idxmax())
print("Highest Sales Region :", region_sales.idxmax())
print("Highest Profit Region :", region_profit.idxmax())
```
